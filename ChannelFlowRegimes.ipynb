```python
import numpy as np
import matplotlib.pyplot as mpl
```

```python
import pathlib
if not pathlib.Path("mpl_utils.py").exists():
  !curl -O https://raw.githubusercontent.com/joaochenriques/MCTE_2022/main/libs/mpl_utils.py &> /dev/null

import mpl_utils as mut
mut.config_plots()

%config InlineBackend.figure_formats = ['svg']
```

```python
def func( ζ2, Fr1 ):
  return 0.5*Fr1**2*(1-1/ζ2**2) + (1-ζ2)
```

```python
Fr = ( 0.05, 0.1, 0.2, 0.3, 0.5, 0.8 )
ζ2 = np.linspace( 0.02, 1, 1000 )

for Fr1 in Fr:
  Δ = func( ζ2, Fr1 )

  mpl.plot( ζ2, Δ, label='Fr$_1$=%.2f' % Fr1 )

mpl.xlabel( 'Fr$_1$ [-]' )
mpl.ylabel( '$\Delta$ [-]' )
mpl.ylim( (0,1) )
mpl.legend();
```
