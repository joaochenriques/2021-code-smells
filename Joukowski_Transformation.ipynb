```python
import numpy as np
import matplotlib.pyplot as mpl
%config InlineBackend.figure_formats = ['svg']
mpl.rcParams["figure.figsize"] = (5, 5)
```

```python
def Joukowski(z,L):
 return z+L/z # asimmetric Joukowski transformation
```

```python
def circle( radius, theta ):
    return radius*np.exp( 1J*theta )

n = 100 # points per side
θ = np.linspace( 0, np.pi*2, 2*n+1 )
```

## Airfoil thickness 

```python
b = 1.0
ε = 0.2

c = np.complex(-b*ε,0.0)
z = circle( b*(1+ε), θ ) + c
w = Joukowski(z,1.0)

mpl.axis('equal')
mpl.grid()

mpl.plot(np.real(z),np.imag(z),'b', label=r'$(\xi,\eta)$')
mpl.plot(np.real(w),np.imag(w),'r', label=r'$(x,y)$')

# draw points
mpl.plot(np.real(z[0]),np.imag(z[0]),'bo')
mpl.plot(np.real(w[0]),np.imag(w[0]),'ro')
mpl.plot(np.real(z[100]),np.imag(z[100]),'bs')
mpl.plot(np.real(w[100]),np.imag(w[100]),'rs')
mpl.plot((0,np.real(c)),(0,np.imag(c)),'g.-')
mpl.plot(np.real(c),np.imag(c),'go')

mpl.legend();
```
